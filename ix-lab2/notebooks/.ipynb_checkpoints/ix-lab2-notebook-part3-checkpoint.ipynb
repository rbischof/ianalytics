## 2.3 Epdemics

#### Exercise 2.9: Simulate an epidemic outbreak

```python
import epidemics_helper
import os
import json
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
with open(os.path.join('../data/nyc_augmented_network.json')) as file:
    data = json.load(file)
G = nx.readwrite.json_graph.node_link_graph(data)
print('Number of nodes:', G.number_of_nodes())
print('Number of edges:', G.number_of_edges())
```

```python
sir = epidemics_helper.SimulationSIR(G, beta=10.0, gamma=0.1, verbose=False)
```

```python
sir.launch_epidemic(source=23654, max_time=100.0)
```

```python
def status(sir):
    r = sir.rec_time.astype(int)
    r_c = np.cumsum(np.bincount(r[r > 0], minlength=100))
    i = sir.inf_time.astype(int)
    i_c = np.cumsum(np.bincount(i[i >= 0], minlength=100))
    h_c = np.zeros(100) + sir.n_nodes - i_c
    return h_c, i_c, r_c

h_c, i_c, r_c = status(sir)
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.title('Evolution of susceptible nodes')
plt.plot(h_c / sir.n_nodes * 100, label='susceptible nodes')
plt.xlabel('Days')
plt.ylabel('Percentage')
plt.legend()
plt.subplot(2,2,2)
plt.title('Evolution of infected nodes')
plt.plot(i_c / sir.n_nodes * 100, label='nodes that have been infected')
plt.plot((i_c - r_c) / sir.n_nodes * 100, label='infected nodes at time t')
plt.xlabel('Days')
plt.ylabel('Percentage')
plt.legend()
plt.subplot(2,2,3)
plt.title('Evolution of recovered nodes')
plt.plot(r_c / sir.n_nodes, label='recovered nodes')
plt.xlabel('Days')
plt.ylabel('Percentage')
plt.legend()
plt.subplot(2,2,4)
plt.title('Superposed percentages of uninfected, infected and recovered nodes')
plt.plot(h_c / sir.n_nodes, label='uninfected nodes')
plt.plot((i_c - r_c) / sir.n_nodes, label='infected nodes')
plt.plot(r_c / sir.n_nodes, label='recovered nodes')
plt.xlabel('Days')
plt.ylabel('Percentage')
plt.legend()
```

```python
print('Time until 60% of population infected:', len(i_c[i_c < sir.n_nodes*.6]), 'days')
print('Time until 60% of population dead:', len(r_c[r_c < sir.n_nodes*.6]), 'days')
```

```python
def visualize_epidemics(sir, src, days, order):
    sir.launch_epidemic(source=src, max_time=days)
    colors = ['green', 'red', 'black']
    classes = ['healthy', 'infected', 'dead']
    xcoord = [[],[],[]]
    ycoord = [[],[],[]]
    for i in range(G.number_of_nodes()):
        x, y = (G.node[i])['coordinates']
        xcoord[sir.status[i]] = np.append(xcoord[sir.status[i]], x / 1e6)
        ycoord[sir.status[i]] = np.append(ycoord[sir.status[i]], y / 1e6)
    plt.scatter(xcoord[order[0]], ycoord[order[0]], c=colors[order[0]], label=classes[order[0]], s=2)
    plt.scatter(xcoord[order[1]], ycoord[order[1]], c=colors[order[1]], label=classes[order[1]], s=2)
    plt.scatter(xcoord[order[2]], ycoord[order[2]], c=colors[order[2]], label=classes[order[2]], s=2)

    plt.title('Epidemic after %d days' % days)
    plt.xlabel('latitude')
    plt.ylabel('longitude')
    plt.legend()
```

```python
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
visualize_epidemics(sir, 23654, 1, [0,1,2])
plt.subplot(2,2,2)
visualize_epidemics(sir, 23654, 3, [0,1,2])
plt.subplot(2,2,3)
visualize_epidemics(sir, 23654, 10, [1,0,2])
plt.subplot(2,2,4)
visualize_epidemics(sir, 23654, 30, [2,1,0])
```

### 2.3.1 Stop the apocalypse!

#### Exercise 2.10: Strategy 1

```python
H_1000 = G.copy()
H_10000 = G.copy()

edges = list(G.edges)
np.random.shuffle(edges)
rem_edges_1000 = edges[:1000]
np.random.shuffle(edges)
rem_edges_10000 = edges[:10000]

for edge in rem_edges_1000:
    H_1000.remove_edge(*edge)
for edge in rem_edges_10000:
    H_10000.remove_edge(*edge)
    
sir_1000 = epidemics_helper.SimulationSIR(H_1000, beta=10.0, gamma=0.1, verbose=False)
sir_10000 = epidemics_helper.SimulationSIR(H_10000, beta=10.0, gamma=0.1, verbose=False)
```

```python
def mean_percentages_after_x_days(G, sir, iterations, days=30):
    stats = np.zeros((iterations,3))
    for i in range(iterations):
        np.random.seed(i)
        src = np.random.choice(range(G.number_of_nodes()))
        sir.launch_epidemic(source=src, max_time=days)
        perc = status(sir)
        for j in range(3):
            stats[i][j] = perc[j][-1]
        
    return np.round(np.mean(stats, axis=0) / sir.n_nodes * 100)
```

```python
normal = mean_percentages_after_x_days(G, sir, 10)
leveraged_1000 = mean_percentages_after_x_days(G, sir_1000, 10)
leveraged_10000 = mean_percentages_after_x_days(G, sir_10000, 10)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)
ind = np.arange(3)
width = .25
n = ax.bar(ind, normal, width)
lev_1000 = ax.bar(ind+width, leveraged_1000, width)
lev_10000 = ax.bar(ind+2*width, leveraged_10000, width)

ax.set_ylabel('Percentage')
ax.set_title('Epidemic distribution after 30 days')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('Succeptible', 'Infected', 'Recovered') )

ax.legend((n, lev_1000, lev_10000),('Normal', '1000 edges removed', "10'000 edges removed"))

plt.show()
```

Removing 10'000 random edges does already have a significant effect on the spread of the epidemic after 30 days.

#### Exercise 2.11: Strategy 2

We need to isolate as many parts of the city as possible. In New York that would most easily be done by blocking the bridges. But how to locate those edges? One idea would be to use the min_edge_cut implemented in networkx. But this method proved to be computationally too heavy for our machines. So we needed to come up with other solutions.

Looking at the map of nodes and edges, we noticed that the bridges over the water have (conveniently) intersections on them, i.e. nodes that are located on bridges have exactly 2 neighbors. 
So we think that by removing the edges from all nodes that have 2 neighbors, we can cut off the bridges and isolate the different islands of New York, which would stop the epidemic from spreading.
In the following method we remove all edges from nodes with len(neighbors) = 2, and if we still have budget left, we continue removing from nodes with len(neighbors) = 3 and so on.

```python
def removeBridgeEdges(G, budget):
    g = G.copy()
    b = min(budget, g.number_of_edges())
    nodes = list(g.nodes)
    target = 2
    
    while b > 0:
        for n in nodes:
            if b <= 0:
                break
            neighbors = list(g.neighbors(n))
            l = len(neighbors)
            if l > 1 and l <= target:
                for neigh in neighbors:
                    g.remove_edge(n, neigh)
                    b -= 1
        target += 1
    return g
```

```python
sir_1000_2 = epidemics_helper.SimulationSIR(removeBridgeEdges(G, 1000), beta=10.0, gamma=0.1, verbose=False)
sir_2500_2 = epidemics_helper.SimulationSIR(removeBridgeEdges(G, 2500), beta=10.0, gamma=0.1, verbose=False)
sir_10000_2 = epidemics_helper.SimulationSIR(removeBridgeEdges(G, 10000), beta=10.0, gamma=0.1, verbose=False)
```

```python
leveraged_1000_2 = mean_percentages_after_x_days(G, sir_1000_2, 10, 30)
leveraged_2500_2 = mean_percentages_after_x_days(G, sir_2500_2, 10, 30)
leveraged_10000_2 = mean_percentages_after_x_days(G, sir_10000_2, 10, 30)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)
ind = np.arange(3)
width = .14
n = ax.bar(ind, normal, width)
lev_1000 = ax.bar(ind+width, leveraged_1000_2, width)
lev_2500 = ax.bar(ind+2*width, leveraged_2500_2, width)
lev_10000 = ax.bar(ind+3*width, leveraged_10000_2, width)

ax.set_ylabel('Percentage')
ax.set_title('Epidemic distribution after 100 days')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('Succeptible', 'Infected', 'Recovered') )

ax.legend((n, lev_1000, lev_2500, lev_10000),('normal', '1000 edges removed', '2500 edges removed', "10'000 edges removed"))

plt.show()
```

When removing 1000 edges we didn't remove all bridges (budget too low). So our method is barely better than without removing edges at all. However, after removing 10'000 edges we successfully isolated the islands, which is visualized in the following plots.

```python
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
visualize_epidemics(sir_10000_2, 23654, 1, [0,1,2])
plt.subplot(2,2,2)
visualize_epidemics(sir_10000_2, 23654, 3, [0,1,2])
plt.subplot(2,2,3)
visualize_epidemics(sir_10000_2, 23654, 10, [1,0,2])
plt.subplot(2,2,4)
visualize_epidemics(sir_10000_2, 23654, 30, [2,1,0])
```
